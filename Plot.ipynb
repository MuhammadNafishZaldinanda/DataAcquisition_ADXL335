```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\Tugas Akhir (Skripsi)\Data\Inner\datainner_1.csv')  
df
```

```python
waktu = df.iloc[:, 0]
amplitude = df.iloc[:, 1]
amplitude1 = df.iloc[:, 2]
amplitude2 = df.iloc[:, 3]
```

```python
plt.figure(figsize=(10,5))
plt.plot(waktu, amplitude, color = "blue", label='Sumbu X')
plt.plot(waktu, amplitude1, color = "green", label='Sumbu Y')
plt.plot(waktu, amplitude2, color = "red", label='Sumbu Z')
plt.xlabel('Time(s)', fontsize=18)
plt.ylabel('Acceleration (g)', fontsize=18)
#plt.ylim(-0.6, 1.8)
#plt.xlim(0, 1)
plt.legend()
#ax = plt.gca()
#ax.set_xticklabels(["{:.1f}".format (x) for x in ax.get_xticks()], fontsize=14)
#ax.set_yticklabels(["{:.1f}".format(y) for y in ax.get_yticks()], fontsize=14) 
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex = True, figsize=(20,7))
ax1.plot(amplitude, color = "blue")
ax1.set_title('Sumbu X')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Acceleration (g)')
ax2.plot(amplitude1, color = "green")
ax2.set_title('Sumbu Y')
ax2.set_xlabel('Time (s)\n')
ax2.set_ylabel('Acceleration (g)')
ax3.plot(amplitude2, color = "red")
ax3.set_title('Sumbu Z')
ax3.set_xlabel('Time (s)')
ax3.set_ylabel('Acceleration (g)')
plt.subplots_adjust(wspace=0.4, hspace=0.7)
```
